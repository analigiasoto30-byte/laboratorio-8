```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {"Mes": ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"],
        "Ventas": [25, 30, 28, 35, 40, 50, 55, 60, 70, 80, 75, 90]}

df_ventas = pd.DataFrame(data)
display(df_ventas.head())
```

Ahora, vamos a realizar algunas operaciones con NumPy para obtener estadísticas básicas de las ventas, como la venta mínima, máxima, promedio y la suma total.

```python
min_ventas = np.min(df_ventas['Ventas'])
max_ventas = np.max(df_ventas['Ventas'])
promedio_ventas = np.mean(df_ventas['Ventas'])
suma_ventas = np.sum(df_ventas['Ventas'])

print(f"Venta mínima: {min_ventas} (miles)")
print(f"Venta máxima: {max_ventas} (miles)")
print(f"Venta promedio: {promedio_ventas:.2f} (miles)")
print(f"Suma total de ventas: {suma_ventas} (miles)")
```

Finalmente, visualicemos las ventas mensuales utilizando Matplotlib para ver la tendencia a lo largo del año.

```python
plt.figure(figsize=(10, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', linestyle='-', color='skyblue')
plt.title('Ventas Mensuales (en miles)')
plt.xlabel('Mes')
plt.ylabel('Ventas (miles)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data_calificaciones = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

df_calificaciones = pd.DataFrame(data_calificaciones)
display(df_calificaciones.head())
```

Ahora, calcularemos las estadísticas básicas (mínima, máxima, promedio) para cada asignatura utilizando NumPy.

```python
asignaturas = ['Matemáticas', 'Ciencias', 'Historia']

for asignatura in asignaturas:
    min_calif = np.min(df_calificaciones[asignatura])
    max_calif = np.max(df_calificaciones[asignatura])
    promedio_calif = np.mean(df_calificaciones[asignatura])

    print(f"--- {asignatura} ---")
    print(f"Calificación mínima: {min_calif}")
    print(f"Calificación máxima: {max_calif}")
    print(f"Calificación promedio: {promedio_calif:.2f}\n")
```

Finalmente, vamos a visualizar las calificaciones de cada estudiante por asignatura utilizando un gráfico de barras para facilitar la comparación.

```python
df_calificaciones.set_index('Estudiante').plot(kind='bar', figsize=(12, 7), colormap='viridis')
plt.title('Calificaciones de Estudiantes por Asignatura')
plt.xlabel('Estudiante')
plt.ylabel('Calificación')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
