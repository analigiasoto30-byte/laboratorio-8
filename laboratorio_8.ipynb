```python
import random

def generar_numero():
    """Genera un número aleatorio entre 1 y 100."""
    return random.randint(1, 100)

def jugar_adivina_el_numero():
    print("¡Bienvenido al juego de Adivina el Número!")
    print("Estoy pensando en un número entre 1 y 100.")

    numero_secreto = generar_numero()
    intentos = 0
    adivinado = False

    while not adivinado:
        try:
            entrada_usuario = input("Adivina el número: ")
            intento_jugador = int(entrada_usuario)
            intentos += 1

            if intento_jugador < numero_secreto:
                print("¡Demasiado bajo! Intenta de nuevo.")
            elif intento_jugador > numero_secreto:
                print("¡Demasiado alto! Intenta de nuevo.")
            else:
                print(f"¡Felicidades! Adivinaste el número {numero_secreto} en {intentos} intentos.")
                adivinado = True
        except ValueError:
            print("Entrada no válida. Por favor, ingresa un número entero.")

# Para jugar, simplemente llama a la función:
jugar_adivina_el_numero()
```

```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        raise ValueError("No se puede dividir por cero.")
    return a / b

def calculadora():
    while True:
        print("\n--- Calculadora Básica ---")
        print("1. Suma")
        print("2. Resta")
        print("3. Multiplicación")
        print("4. División")
        print("5. Salir")

        opcion = input("Selecciona una opción (1-5): ")

        if opcion == '5':
            print("¡Hasta luego!")
            break

        if opcion in ('1', '2', '3', '4'):
            try:
                num1 = float(input("Ingresa el primer número: "))
                num2 = float(input("Ingresa el segundo número: "))

                if opcion == '1':
                    print(f"Resultado: {num1} + {num2} = {suma(num1, num2)}")
                elif opcion == '2':
                    print(f"Resultado: {num1} - {num2} = {resta(num1, num2)}")
                elif opcion == '3':
                    print(f"Resultado: {num1} * {num2} = {multiplicacion(num1, num2)}")
                elif opcion == '4':
                    print(f"Resultado: {num1} / {num2} = {division(num1, num2)}")

            except ValueError as e:
                print(f"Error: {e}")
            except ZeroDivisionError:
                print("Error: No se puede dividir por cero.")
            except Exception as e:
                print(f"Ocurrió un error inesperado: {e}")
            finally:
                # En este caso, 'continue' ya está implícito en el bucle while True
                # si no hay un break. Se usa para volver al inicio del bucle.
                continue
        else:
            print("Opción no válida. Por favor, intenta de nuevo.")

# Para usar la calculadora, simplemente llama a la función:
calculadora()
```

```python
tareas = []

def agregar_tarea(tarea):
    tareas.append(tarea)
    print(f"Tarea '{tarea}' agregada.")

def eliminar_tarea(indice):
    try:
        if 0 <= indice < len(tareas):
            tarea_eliminada = tareas.pop(indice)
            print(f"Tarea '{tarea_eliminada}' eliminada.")
        else:
            print("Índice de tarea no válido.")
    except ValueError:
        print("Por favor, ingresa un número válido para el índice.")

def mostrar_tareas():
    if not tareas:
        print("No hay tareas en la lista.")
    else:
        print("\n--- Tu Lista de Tareas ---")
        for i, tarea in enumerate(tareas):
            print(f"{i}. {tarea}")
        print("-------------------------")

def lista_de_tareas():
    while True:
        print("\n--- Menú de Tareas ---")
        print("1. Agregar tarea")
        print("2. Eliminar tarea")
        print("3. Mostrar tareas")
        print("4. Salir")

        opcion = input("Selecciona una opción (1-4): ")

        if opcion == '1':
            tarea_nueva = input("Ingresa la nueva tarea: ")
            agregar_tarea(tarea_nueva)
        elif opcion == '2':
            mostrar_tareas()
            try:
                indice_eliminar = int(input("Ingresa el número de la tarea a eliminar: "))
                eliminar_tarea(indice_eliminar)
            except ValueError:
                print("Entrada no válida. Por favor, ingresa un número entero.")
        elif opcion == '3':
            mostrar_tareas()
        elif opcion == '4':
            print("¡Hasta pronto!")
            break
        else:
            print("Opción no válida. Por favor, intenta de nuevo.")

# Para usar la lista de tareas, simplemente llama a la función:
lista_de_tareas()
```
